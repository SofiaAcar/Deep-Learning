# Dataset Info

Appliance Energy Usage: A Multivariate Time Series Forecasting Example

Experimental data used to create regression models of appliances energy use in a low energy building.
Data Set Information:

The data set is at 10 min for about 4.5 months. The house temperature and humidity conditions were monitored with a ZigBee wireless sensor network. Each wireless node transmitted the temperature and humidity conditions around 3.3 min. Then, the wireless data was averaged for 10 minutes periods. The energy data was logged every 10 minutes with m-bus energy meters. Weather from the nearest airport weather station (Chievres Airport, Belgium) was downloaded from a public data set from Reliable Prognosis (rp5.ru), and merged together with the experimental data sets using the date and time column. Two random variables have been included in the data set for testing the regression models and to filter out non predictive attributes (parameters).
Original source of the dataset:

http://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction

# Data Pre-Processing

## Feature Engineering
- Let's reduce the number of feature
- Let's resample the dataset by hours

__!!!__ Categoric sütunlar olsaydi get_dummies veya ordinal encoder islemi yapmamiz gerekebilirdi. Bazi notebook' larda get_dummy yapilmasi gereken sütunlarin yapilmadigi orneklerle karsilasabiliriz. Bazen yapilmamasi daha iyi sonuclar verebiliyor, denemek gerekir. __!!!__

Cok fazla sütun oldugu icin analizi kolaylastirmak adina feature enginering yapacagiz :

```python
df.columns
```

Anlasilmasi daha kolay olsun diye T6 ve RH_6 sütununu yeniden isimlendirdik :

```python
df=df.rename(columns = {'T6':'T_outside', 'RH_6':'RH_outside'})
```

Sicaklik degerlerinin ortalamasini alarak bunlari tek bir sütun haline getirecegiz :

```python
df["T_inside"] = (df["T1"] + df["T2"] + df["T3"] + df["T4"] + df["T5"] + df["T7"] + df["T8"] + df["T9"])/8
```

Nem degerlerinin ortalamasini alarak bunlari da tek bir sütun haline getirecegiz :

```python
df["RH_inside"] = (df["RH_1"] + df["RH_2"] + df["RH_3"] + df["RH_4"] + df["RH_5"] + df["RH_7"] + df["RH_8"] + df["RH_9"])/8
```

Ihtiyacimiz olamayan feature' lari datadan attik :

```python
drop_list = ['T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3', 'T4', 'RH_4', 'T5', 'RH_5',
             'T7', 'RH_7', 'T8', 'RH_8', 'T9', 'RH_9', 'T_out', 'RH_out', 'rv1', 'rv2']
df.drop(drop_list, axis = 1, inplace = True)
```

__Appliances,__ cihazlarin enerji harcama durumu; __lights,__ isiklarin enerji harcama durumu.

```python
df.info()
```

Datadaki veriler 10 dk da bir ölcüldügü icin bir saatte 6, bir günde ise 24*6=144 sequence uzunlugumuzun olmasi gerekir. Bu cok uzun bir sequence olur, LSTM bu kadar cok bilgiyi aklinda tutamaz, önemli kisimlari da atlayabilir. 

Datamizi ikiye ayirdik. Ilk degiskene Appliances ve lights' i (enerji harcamalari) ikinci degiskene ise Appliances ve ligths disindaki feature' lari (sicaklik) tanimladik :

```python
df_a = df[["Appliances", "lights"]]                      # Enerji miktarlari
df_b = df.drop(["Appliances", "lights"], axis = 1)       # Sicaklik miktarlari
```

Timeseries datalarinda kullanilan bir fonksiyon olan __resample__ ile dakikayi saate cevirdik. resample isleminden sonra bir aggregation kullanmamiz gerekir cunku 6 tane veri elde edildi. Ilk degiskendeki enerji harcamalarinin bir saat icinde cikan 6 deger toplandi. Ikinci degiskendeki sicaklik degerlerinin ise bir saat icinde cikan 6 degerinin ortalamasi alindi :

```python
df_a = df_a.resample("H").sum()
df_b = df_b.resample("H").mean()
```

Bu iki data concat islemi ile tekrar birlestirildi :

```python
df = pd.concat([df_a, df_b], axis =1)
df.head()
```

```python
df.shape
```

Let's also round off the data, to one decimal point precision, otherwise this may cause issues with our network (we will also normalize the data anyways, so this level of precision isn't useful to us)

```python
df = df.round(2)    # Virgullu degerlerden kurtulmak icin, virgulden sonraki 2 basamagi al.
```

Yeni datamiz ile heatmap grafigine tekrar baktik. Baslangicta gördügümüz cok yüksek corr iliskileri artik yok :

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True);
```

__Appliances__ sütununa yogunlasmak istedigimiz icin bu sütunu grafikte inceledik. Saatlik olarak elde edilmis 5-6 aylik bir veri var. grafikteki veriler cok yogun oldugu icin yorumlanmasi zor. Bu yüzden asagida 3 günlük bir veri secip bunu inceledik :

Her gun ogleden sonra 16.00 civarinda enerji harcamalarinda bir patlama goruluyor. Ikinci gun sabah saatlerinde de ufak bir patlama var. Bu verilerle model kurmak zor gibi görünüyor :

### Train Test Split

```python
len(df)
```

Saatlik bir verimiz oldugu icin length=24 alacagiz.

```python
# How many rows per day?
24
```

Test datasi icin 7 gun ayirmaya karar verdik. 7*24=168 adet test datamiz olacak :

```python
test_days = 7
```

```python
test_ind = test_days*24
```

```python
test_ind
```

Train datasi icin tum datanin ilk kismini, test datasi icin ise son 168 satiri ayirdik :

```python
train = df.iloc[:-test_ind]
test = df.iloc[-test_ind:]
```

```python
train.shape, test.shape
```

### Scale Data

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

## Create the Model

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

Datamiz saatlik oldugu icin 24 saatte bir periyot tamamlanacak. Bu yuzden length=24 olarak belirledik. batch_size' i da bir degiskene atadiktan sonra generator icine datalarimizi ve parametreleri vererek datamizi modele uygun hale getirdik :

```python
length = 24 # Length of the output sequences (in number of timesteps)
batch_size = 1 #Number of timeseries samples in each batch
generator = TimeseriesGenerator(train_scaled, train_scaled, length=length, batch_size=batch_size)
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length=length, batch_size=batch_size)
```

3122-24=3098 generator (Son 24 satirda islem yapamayacak) (train icin)

168-24=168 generator (Test icin)

```python
len(train_scaled), len(generator), len(test_scaled), len(validation_generator)
```

```python
n_features = train_scaled.shape[1]
```

__model.add(LSTM(100,input_shape=(length, n_features))) -->__ length=24, feature=10 

Bir hidden layer ile LSTM modelimizi kurduk. 

__model.add(Dense(n_features)) -->__ Simdiye kadarki tüm calismalarimizda output' a 1 nöron koymustuk, burda ise feature sayisi kadar tanimladik. TimeseriesGenereator, her bir sütunu kendi icinde degerlendirip o time' e göre bir prediction yapar ve bunu her feature icin yapar. Sadece Appliances üzerine calisip, diger feature' larin Appliances üzerine olan etkisini inceleyen bir model olusturmaz. Kac feature varsa hepsi icin tahmin yapar. Her feature target gibidir ve hepsi icin regression yapacak. Bu yüzden TimeseriesGenerator, gercek hayat multivariate datalarla calisma icin uygun degildir.

```python
model = Sequential()

model.add(LSTM(100,input_shape=(length, n_features)))

model.add(Dense(n_features))

model.compile(optimizer='adam', loss='mse')
```

```python
early_stop = EarlyStopping(monitor='val_loss',patience=3)
```

```python
model.fit_generator(generator,
                    epochs=10,
                    validation_data=validation_generator,
                    callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

## Evaluate on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

Train datasi 10 feature ve 24 satirdan olusan datalar halinde :

```python
train_scaled
```

Onceki notebooktaki ayni fonksiyonu kullanarak train datasinin son 24 datasi ile forecast islemlerimizi yaptik  ve prediction' larimizi aldik :

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch)
    
    # store prediction
    predictions_scaled.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis=1)
```

prediction islemi sonunda kac feature varsa (10) o kadar sonuc üretildi (Onceki calismada bir sonuc üretilmisti, burada her feature icin birer sonuc üretildi). legth=24 oldugu icin her feature icin 24 adet tahmin yapildi :

```python
predictions_scaled
```

Train datasinin bittigi yerde test datasi basliyor. Yukaridaki islemde train datasindan sonra 24 satir tahmin yaptirdik. Bu degerler test datasinin ilk 24 degerine karsilik geliyor. Prediction' larimizin skorlarina bakabilmek icin test datasinin ilk 24 verisini ve prediction' lari eval_metric icine verdik. Buradaki skor, 10 feature' daki gercek degerler ile bunlara karsilik bulunan 10 feature' lik prediction' larin karsilastirilmasi sonucu elde edilen ortalama bir r2 skoru :

```python
eval_metrics(test_scaled[:length], predictions_scaled)
```

## Inverse Transformations and Compare

Gercek DataFrame' deki her feature ile yaptigimiz prediction' in feature' larini tek tek karsilastiracagiz. Bunun icin tüm prediction' lari inverse_transform ile scale edilmis halden gercek degerlerine cevirdik ve DataFrame' e donusturduk. Feature isimleri olarak da gercek df' in column isimlerini verdik (df veya test isimleri verilebilir, ikisi de ayni) :

```python
predictions = scaler.inverse_transform(predictions_scaled)
predictions = pd.DataFrame(data=predictions,columns=test.columns)
predictions
```

Prediction degerlerini test datasinin ilk 24 degeri ile karsilastiracagiz. Bu yuzden test datasindaki ilk 24 satiri bir degiskene attik :

```python
test_new = test[:length]
```

Bir for döngüsü kurarak eval_metric icine yeni olusturdugumuz test datasini ve predictionlari verdik. Böylece test'n ve prediction' in ayni sütunlarini alacak ve karsilastiracak. 10 sütun icin 10 ayri skor bulundu ve skorlar genel olarak anlamsiz cikti :

```python
for i in test_new.columns:
    print(i)
    eval_metrics(test_new[i], predictions[i])
    print("--"*15)
```

Daha iyi bir kiyaslama yapabilmek icin buldugumuz degerlerin grafiklerini cizdirdik. Genel olarak gercek degerler ile prediction degerleri birbirinden oldukca farkli cikti. Her bir feature' i etkileyen baska faktorler de var, biz bu faktorleri hesaba katmadan sadece zamana gore bir degerlendirme yaptigimiz icin mantikli bir sonuc elde edemedik. Datanin indexinde time olmasi, Timeseries datamiz oldugu anlamina gelmiyor :

```python
index = 0
plt.figure(figsize = (20, 30))

for i in test.columns:
    index += 1
    test_eva = pd.DataFrame(test_new[i])                    # test_new' in ilgili su1tunu ile DataFrame olustur.
    test_eva["predictions"] = predictions[i].values         # Olusan DataFrame' e ilgili prediction sütununu ekle (Her seferinde 2 feature' li bir df olusacak)
    plt.subplot(5, 2, index)                                # 5*2=10 grafik, index her seferinde 1 artacak.
    sns.lineplot(x = test_eva.index, y = test_eva[i])
    sns.lineplot(x = test_eva.index, y = test_eva.predictions)
    plt.xticks(rotation=90);
```
